```python
import requests
import os

# Step 1: Fetch the webpage
url = "https://quotes.toscrape.com/"
response = requests.get(url)

# Step 2: Make sure the folder exists
folder_name = "web_scrapping"
os.makedirs(folder_name, exist_ok=True)

# Step 3: Define the file path
file_path = os.path.join(folder_name, "quotes.html")

for i in range (1,5):
  with open(file_path, "w", encoding="utf-8") as file:
    file.write(response.text)

print(f"Data saved successfully at {file_path}")
```

```python
from bs4 import BeautifulSoup
```

```python
with open(file_path, "w+", encoding="utf-8") as file:
    file.write(response.text)
    file.seek(0)  # move cursor back to start
    content = file.read()
```

```python
soup=BeautifulSoup(content,"html.parser")
```
